## Numpy

```python
import numpy as np
data = np.array([15, 16, 18, 19, 22, 24, 29, 30, 34])
print("Mean:\t\t\t", data.mean())
print("Median:\t\t\t", np.median(data))
print("Distance from mean: \t", abs(data.mean() - data))
print("Array length:\t\t", len(data))
```

$8^2+7^2+5^2+4^2+1^2+1^2+6^2+7^2+11^2 = \\ 64+49+25+16+1+1+36+49+121=362\\$
$132/9 = 40.22 \\ \sqrt{40.22} = 6.342$

```python
print("Std is: \t", np.sqrt(sum(abs(data - data.mean())**2)/len(data)))
```

```python
print("mean: \t\t\t\t", np.mean(data))
print("median: \t\t\t", np.median(data))
print("50th percentile (median): \t", np.percentile(data, 50))
print("25th percentile: \t\t", np.percentile(data, 25))
print("75th percentile: \t\t", np.percentile(data, 75))
print("standard deviation: \t\t", np.std(data))
print("Variance: \t\t\t", np.var(data))
```

## Pandas

```python
import pandas as pd
pd.options.display.max_columns = 6
df = pd.read_csv('https://sololearn.com/uploads/files/titanic.csv')
print(df.head())
```

```python
print(df.describe())
```

#### Col is Series type, df is DataFrame type

```python
col = df['Fare']
print(col)
```

#### Smaller DataFrame

```python
small_df = df[['Age', 'Sex', 'Survived']]
print(small_df.head())
```

#### Making new columns

```python
df['male'] = df['Sex'] == 'male'
print(df.head())
```

## Numpy Basics

#### Pandas DataFrame to Numpy 1dArray

```python
print(df['Fare'].values)
```

#### Pandas DataFrame to Numpy 2dArray

```python
print(df[['Pclass', 'Fare', 'Age']].values)
```

#### Numpy shape

```python
arr = df[['Pclass', 'Fare', 'Age']].values
print(arr.shape)
```

## Masks

```python
mask = df['Pclass'] == 1
print(df[mask])
```
